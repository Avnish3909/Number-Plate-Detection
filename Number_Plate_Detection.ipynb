```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

video = cv2.VideoCapture('/content/new_york_city_-_1044 (360p).mp4')
car_cascade = cv2.CascadeClassifier('/content/HaarCascade_cars.xml')

while video.isOpened():
    ret, frame= video.read()
    if (type(frame) == type(None)):
        break
    frame= cv2.resize(frame, None, fx=0.15, fy=0.15, interpolation= cv2.INTER_LINEAR)
    gray= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cars = car_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=4)
    for (x,y,w,h) in cars:
        cv2.rectangle(frame,(x,y),(x+w,y+h), (0,255,255), 2)
        cv2_imshow(frame)
    if cv2.waitKey(1) == 13:
        break
video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

carplate_cascade = cv2.CascadeClassifier('/content/haarcascade_russian_plate_number.xml')

img = cv2.imread('/content/car-1560628_1280.jpg')

carplates = carplate_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5)

for x,y,w,h in carplates:
	cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 5)

scale_factor = 1.2
width = int(img.shape[1] * scale_factor / 100)
height = int(img.shape[0] * scale_factor / 100)
dim = (width, height)
cv2_imshow(img)
cv2.imwrite('op2.jpg', img)
cv2.waitKey(0)
```
